# Gamma Alice29

```python
import math
from math import log
```

# Convert to ByteList

```python
alice = open("files/alice29.txt", "r")

alice_str = alice.read()
encoded = alice_str.encode(encoding="utf-8")

byte_list=[]

for l in encoded:
    byte_list.append(l)
```

# Gamma Encoding

```python
log2 = lambda x: log(x, 2)

def Unary(x):
    return (x-1)*'0'+'1'

def Binary(x, l = 1):
    s = '{0:0%db}' % l
    return s.format(x)

def Elias_Gamma(x):
	array = []
	for num in x:
		if(num == 0):
			return '0'
		else:
			n = 1 + int(log2(num))
			b = num - 2**(int(log2(num)))

			l = int(log2(num))
			sum = Unary(n) + Binary(b, l)
			array.append(sum) 
	return(array)

result = Elias_Gamma(byte_list)

B = [str(k) for k in result]
C="[1,0]"+",".join(B)

open("files/encoded/alice29_encoded_gamma.txt", "w").write(C)
```
