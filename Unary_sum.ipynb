# Unary Sum

# Convert to ByteList

```python
sum = open("files/sum", "rb")

sum_str = sum.read()
byte_list=[]

for s in sum_str:
	byte_list.append(s)
```

# Unary Encoding

```python
N = byte_list
A = []

for i in N:
		B = []
		for n in range(i):
				B.append(1)
		B.append(0)
		A.append(B)

B = [str(k) for k in A]

C = ",".join(B)

open("files/encoded/sum_encoded_unary.txt", "w").write(C)
```

# Unary Decoding

```python
code =  C.replace(",", "").replace("[", "").replace("]", "").replace(" ", "")
count = 0
array = []
for i in code:
    for j in i:
        if j == "1":
            count += 1
        else:
            array.append(count)
            count = 0
to_byte=[]
to_byte.append(array)
```

# Writing file

```python
array_list=[]

for value in to_byte:
  array_list.append(value)

open("files/decoded/sum_decoded_unary", "wb").write(bytes(array_list[0]))
```
