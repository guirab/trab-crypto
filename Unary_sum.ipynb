# Unary Sum

# Convert to ByteList

```python
sum = open("sum", "rb")

sum_str = sum.read()

byte_list=[]

for s in sum_str:
	byte_list.append(s)
```

# Unary Encoding

```python
N = byte_list
A = []

for i in N:
		B = []
		for n in range(i):
				B.append(1)
		B.append(0)
		A.append(B)

B = [str(k) for k in A]

C = ",".join(B)

# open("sum_encoded_unary", "wb").write(C)
```

# Unary Decoding

```python
code =  C.replace(",", "").replace("[", "").replace("]", "").replace(" ", "")
count = 0
array = []
for i in code:
    for j in i:
        if j == "1":
            count += 1
        else:
            array.append(count)
            count = 0
to_byte=[]
for byte in array:
    to_byte.append(byte.to_bytes(1, byteorder='big'))
```

# Writing file

```python
array_list=""
for value in to_byte:
  b = str(value)
  array_list+=b

print(array_list)
# open("sum_decoded_unary", "wb").write(array_list)
```
