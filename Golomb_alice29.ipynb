# Golomb Alice29

```python
import math
```

# Convert to ByteList

```python
alice = open("alice29.txt", "r")

alice_str = alice.read()
encoded = alice_str.encode(encoding="utf-8")

byte_list=[]

for l in encoded:
    byte_list.append(l)
```

# Golomb Encoding

```python
M = int(input("Digite o valor do divisor: "))

res = []
for number in byte_list:
    ans = []
    q = number//M
    r = number % M

    quo ='0'*q+'1'

    b = math.floor(math.log2(M))
    k = 2**(b + 1)-M
    if r < k:
        rem = bin(r)[2:]
        l = len(rem)
        if l<b:
            rem = '0'*(b-l)+rem
    else:
        rem = bin(r + k)[2:]
        l = len(rem)
        if l<b + 1:
            rem = '0'*(b + 1-l)+rem
    ans.append(quo + rem)
    res.append(ans)
B = [str(k) for k in res]
C=",".join(B)
# print(C)
# open("alice29_encoded_golomb.txt", "w").write(C)
```

# Golomb Decoding

```python
code =  C.replace(",", "").replace("[", "").replace("]", "").replace(" ", "").replace("'", "")

def reccurent():
    nums = []
    number = ""
    
    offset = 0
    for index, i in enumerate(code):

        if index < offset:
            continue

        number += i
        if i != "0":

            if len(number) == 1:
                nums.append(number)
                offset = index + 1
            else:
                length = len(number) * 2 - 1
                word = code[offset: offset + length]
                nums.append(word)
                offset += length
            number = ""

    print(nums)
    
reccurent()
```
