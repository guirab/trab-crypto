# Unary Alice29

# Convert to ByteList

```python
alice = open("files/alice29.txt", "r")

alice_str = alice.read()
encoded = alice_str.encode(encoding="utf-8")

byte_list=[]

for l in encoded:
    byte_list.append(l)
```

# Unary Encoding

```python
N = byte_list
A = []

for i in N:
		B = []
		for n in range(i):
				B.append(1)
		B.append(0)
		A.append(B)

B = [str(k) for k in A]
C = "[3,0]"+",".join(B)

open("files/encoded/alice29_encoded_unary.txt", "w").write(C)
```

# Unary Decoding

```python
code =  C.replace(",", "").replace("[", "").replace("]", "").replace(" ", "")

count = 0
array = []
for i in code:
    for j in i:
        if j == "1":
            count += 1
        else:
            array.append(count)
            count = 0
to_byte=[]
for byte in array:
    to_byte.append(byte.to_bytes(1, byteorder='big'))
```

# Writing file

```python
array_list=""

for value in to_byte:
  b = str(value.decode("utf-8"))
  array_list+=b

open("files/decoded/alice29_decoded_unary.txt", "w").write(array_list)
```
