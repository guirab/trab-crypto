# Delta Encoding

```python
def delta_encode(list):
  last = 0
  array = []
  for i in list:
    if last == 0:
      array.append(i)
      last = i
    else:
      array.append(last - i)
      last = last - i
  return array

code = delta_encode(byte_list)
```
