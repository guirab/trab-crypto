# Delta Alice29

# Convert to ByteList

```python
alice = open("files/alice29.txt", "r")

alice_str = alice.read()
encoded = alice_str.encode(encoding="utf-8")

byte_list=[]

for l in encoded:
    byte_list.append(l)
```

# Delta Encoding

```python
def delta_encode(list):
  last = 0
  array = []
  for i in list:
    if last == 0:
      array.append(i)
      last = i
    else:
      array.append(last - i)
      last = last - i
  return array

code = delta_encode(byte_list)

open("files/encoded/alice29_encoded_delta.txt", "w").write(str(code))
```

# Delta Decoding

```python
def delta_decode(lista):
  last = 0
  array = []
  lista = list(lista)
  for i in lista:
    if last == 0:
      array.append(i)
      last = i
    elif i == 0:
      array.append(last)
      last = 0
    elif i < 0:
      array.append(-i + last)
      last = i
    else:
      array.append(i - last)
      last = i - last
  return array

to_byte=[]
for byte in delta_decode(code):
    to_byte.append(byte.to_bytes(1, byteorder='big'))
```

# Writing file

```python
array_list=""

for value in to_byte:
  b = str(value.decode("utf-8"))
  array_list+=b

open("files/decoded/alice29_decoded_delta.txt", "w").write(array_list)
```
