# Fibo Alice29

# Convert to ByteList

```python
alice = open("files/alice29.txt", "r")

alice_str = alice.read()
encoded = alice_str.encode(encoding="utf-8")

byte_list=[]

for l in encoded:
    byte_list.append(l)
```

# Def Fibo

```python
def fibonacci(n):
	a = 0
	b = 1
	if n < 0:
		return
		
	elif n == 0:
		return 0
	
	elif n == 1:
		return b
	else:
		for i in range(1, n):
			c = a + b
			a = b
			b = c
		return b
```

# Fibo Encoding

```python
array = []

def encode(x):
    for number in x:
        retv = []
        R = []
        i = 0
        while fibonacci(i) < number:
            i += 1
        while number > 0:
            if fibonacci(i) <= number:
                R.append(i)
                number-= fibonacci(i)
            i -= 1
        for i in range(max(R) + 1):
            if i in R:
                retv.append(1)
            else:
                retv.append(0)
        retv.append(1)
        array.append(retv)
    return(array)

code = encode(byte_list)
B = [str(k) for k in code]
C="[2,0]"+",".join(B)

open("files/encoded/alice29_encoded_fibo.txt", "w").write(C)
```

# Fibo Decoding

```python
decoded_array = []

def decode(x):
    for l in x:
        l = str(l).replace("[", "").replace("]", "").replace(" ", "").replace("'", "").replace(",", "")
        retv = 0
        for i in range(len(l) - 1):
            if l[i] == '1':
                retv += fibonacci(i)
        decoded_array.append(retv)
    return(decoded_array)

to_byte=[]
for byte in decode(code):
    to_byte.append(byte.to_bytes(1, byteorder='big'))
```

# Writing file

```python
array_list=""

for value in to_byte:
  b = str(value.decode("utf-8"))
  array_list+=b

open("files/decoded/alice29_decoded_fibo.txt", "w").write(array_list)
```
