# Delta Sum

# Convert to ByteList

```python
sum = open("files/sum", "rb")

sum_str = sum.read()
byte_list=[]

for s in sum_str:
	byte_list.append(s)
```

# Delta Encoding

```python
def delta_encode(list):
  last = 0
  array = []
  for i in list:
    if last == 0:
      array.append(i)
      last = i
    else:
      array.append(last - i)
      last = last - i
  return array

code = delta_encode(byte_list)
B = [str(k) for k in code]
C="[4]"+",".join(B)

open("files/encoded/sum_encoded_delta.txt", "w").write(C)
```

# Delta Decoding

```python
def delta_decode(lista):
  last = 0
  array = []
  lista = list(lista)
  for i in lista:
    if last == 0:
      array.append(i)
      last = i
    elif i == 0:
      array.append(last)
      last = 0
    elif i < 0:
      array.append(-i + last)
      last = i
    else:
      array.append(last - i)
      last = i
  return array

to_byte=[]
to_byte.append(delta_decode(code))
```

# Writing file

```python
array_list=[]

for value in to_byte:
  array_list.append(value)

open("files/decoded/sum_decoded_delta", "wb").write(bytes(array_list[0]))
```
