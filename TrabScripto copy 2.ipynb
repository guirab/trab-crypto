```python
alice = open("alice29.txt", "r")

alice_str = alice.read()
encoded = alice_str.encode(encoding="utf-8")
byte_list=[]
# print(encoded)

for l in encoded:
    byte_list.append(l)

# print(byte_list)
```

## Convert into bits

```python
# Unary code encoding
N = byte_list
A = []

for i in N:
		B = []
		for n in range(i):
				B.append(1)
		B.append(0)
		A.append(B)

B = [str(k) for k in A]

C = ",".join(B)

# print("Unary code: ", C)
```

```python
code =  C.replace(",", "").replace("[", "").replace("]", "").replace(" ", "")
count = 0
array = []
for i in code:
    for j in i:
        if j == "1":
            count += 1
        else:
            array.append(count)
            count = 0
# print("decoded number is: ", array)
to_byte=[]
for byte in array:
    to_byte.append(byte.to_bytes(1, byteorder='big'))


# def byter(list):
#     return list.to_bytes(1, byteorder='big')

# to_byte = map(byter,array)
# print("Unary decode in bytes: ", to_byte)
```

```python
array_list=""

for value in to_byte:
  b = str(value.decode("utf-8"))
  array_list+=b

open("alice29_decoded_unary.txt", "w").write(array_list)
```
