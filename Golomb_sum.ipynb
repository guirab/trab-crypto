# Golomb Sum

```python
import math
```

# Convert to ByteList

```python
sum = open("files/sum", "rb")

sum_str = sum.read()
byte_list=[]

for s in sum_str:
	byte_list.append(s)
```

# Golomb Encoding

```python
M = int(input("Digite o valor do divisor: "))

res = []
for number in byte_list:
    ans = []
    q = number//M
    r = number % M

    quo ='0'*q+'1'

    b = math.floor(math.log2(M))
    k = 2**(b + 1)-M
    if r < k:
        rem = bin(r)[2:]
        l = len(rem)
        if l<b:
            rem = '0'*(b-l)+rem
    else:
        rem = bin(r + k)[2:]
        l = len(rem)
        if l<b + 1:
            rem = '0'*(b + 1-l)+rem
    ans.append(quo + rem)
    res.append(ans)
B = [str(k) for k in res]
C="[0,"+str(M)+"]"+",".join(B)

open("files/encoded/sum_encoded_golomb.txt", "w").write(C)
```

# Golomb Decoding

```python
code =  C.replace(",", "").replace("[", "").replace("]", "").replace(" ", "").replace("'", "")

def decode(x):
    num=0 
    for i in range(len(x)):
        num+=(int(x[len(x)-1-i])*(math.pow(2,i))) 
    return num 

i=math.floor(math.log(M,2))
d=math.pow(2,i+1)-M

array = []

p2=0 
l=1 
while(p2<len(code)):
    t=0 
    flag=0 
    r=[] 
    k=i 
    q=0 
    for p in range(p2,len(code)):
        if(code[p]=='0' and flag==0):
            t+=1 
            continue 
        if(code[p]=='1' and flag==0):
            q=t 
            flag=1 
            continue 
        r.append(code[p]) 
        k-=1 
        if(k==0):
            rnum=decode(r) 
            if(rnum<d):
                p2=p+1 
                break 
        if(k==-1):
            rnum=decode(r) 
            rnum=rnum-d 
            p2=p+1 
            break 
    array.append(int(q*M+rnum))
    l=0 

to_byte=[]
to_byte.append(array)
```

# Writing file

```python
array_list=[]

for value in to_byte:
  array_list.append(value)

open("files/decoded/sum_decoded_golomb", "wb").write(bytes(array_list[0]))
```
