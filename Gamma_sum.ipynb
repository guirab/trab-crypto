# Gamma Sum

```python
import math
from math import log
```

# Convert to ByteList

```python
sum = open("files/sum", "rb")

sum_str = sum.read()
byte_list=[]

for s in sum_str:
	byte_list.append(s)
```

# Gamma Encoding

```python
log2 = lambda x: log(x, 2)

def Unary(x):
    return (x-1)*'0'+'1'

def Binary(x, l = 1):
    s = '{0:0%db}' % l
    return s.format(x)

def Elias_Gamma(x):
	array = []
	for num in x:
		if(num == 0):
			array.append('0')
		else:
			n = 1 + int(log2(num))
			b = num - 2**(int(log2(num)))

			l = int(log2(num))
			sum = Unary(n) + Binary(b, l)
			array.append(sum) 
	return(array)

result = Elias_Gamma(byte_list)

B = [str(k) for k in result]
C=",".join(B)
open("files/encoded/sum_encoded_gamma.txt", "w").write(C)
```

# Gamma Decoding

```python
def Elias_Gamma_Decoding(x):
	array = []
	for i in x:
		K = 0
		i = list(i)
		while True:
			if len(i) == 1:
				break
			if not i[K] == "0":
				break
			K = K + 1
		i = i[K:2*K+1]

		n = 0
		i.reverse()

		for j in range(len(i)):
			if i[j] == '1':
				n = n+math.pow(2, j)
		array.append(int(n))
	return array

decoded = Elias_Gamma_Decoding(result)

to_byte=[]
to_byte.append(decoded)
```
